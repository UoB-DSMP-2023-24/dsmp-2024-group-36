```python
import pandas as pd
csv_file_path = 'fake_transactional_data_24.csv'

df = pd.read_csv(csv_file_path)
print(df)
df.isnull().sum()
```

```python
df_filtered = df[df['to_randomly_generated_account'].str.contains(r'\D', regex=True)]
unique_shops = df_filtered['to_randomly_generated_account'].unique()
unique_shops
```

```python
# 筛选与A_CAFE相关的交易
a_cafe_transactions = df[df['to_randomly_generated_account'] == 'A_CAFE']

# 显示筛选结果
print(a_cafe_transactions)
```

```python
# 初始化一个空列表来存储每个商户的统计结果
shop_summaries = []

#计算每个商户的总交易额，平均交易额与总交易次数
for shop in unique_shops:
    shops_all_transactions = df_filtered[df_filtered['to_randomly_generated_account'] == shop]
    
    total_amount = shops_all_transactions['monopoly_money_amount'].sum()
    average_amount = shops_all_transactions['monopoly_money_amount'].mean()
    transaction_count = len(shops_all_transactions)
    
    shop_summaries.append({
        'shops': shop,
        'Total_Amount': total_amount,
        'Average_Amount': average_amount,
        'Transaction_Count': transaction_count
    })

# 创建DataFrame
shops_analysis = pd.DataFrame(shop_summaries)

shops_analysis
```

```python
# 排序
shops_analysis_sorted = shops_analysis.sort_values(by='Average_Amount', ascending=False).reset_index(drop=True)

# 分级
shops_analysis_sorted['Consumption_Level'] = 'Standard Spending'
shops_analysis_sorted.loc[shops_analysis_sorted['Average_Amount'] > 100, 'Consumption_Level'] = 'High Spending'
shops_analysis_sorted.loc[(shops_analysis_sorted['Average_Amount'] > 25) & (shops_analysis_sorted['Average_Amount'] <= 100), 'Consumption_Level'] = 'Medium Spending'

#保存并展示
shops_analysis_sorted.to_csv('shops_analysis_with_levels.csv', index=False)
shops_analysis_sorted
```
